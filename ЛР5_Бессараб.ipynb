Для датасету px.data.iris() побудувати наступні графіки.  

Графіки мають назву, що вирівнювання за центром.

1.Побудуйте 3D-графік розподілу даних для датасету Iris з використанням властивостей "sepal_length," "sepal_width," та "petal_length." Кожна точка має бути розфарбована відповідно до класу квітки (setosa, versicolor, virginica).

```python
import plotly.express as px

df = px.data.iris()

fig = px.scatter_3d(
    df,
    x='sepal_length',
    y='sepal_width',
    z='petal_length',
    color='species',
    title='3D Розподіл даних Iris за ознаками чашолистка та пелюстки'
)

fig.update_layout(
    title={
        'x': 0.5,
        'xanchor': 'center'
    }
)

fig.show()
```

2.Побудуйте гістограму для розподілу значень "petal_width" для різних видів квіток в датасеті Iris.

```python
import plotly.express as px

df = px.data.iris()

fig = px.histogram(
    df,
    x='petal_width',
    color='species',
    barmode='overlay',
    nbins=20,
    title='Розподіл ширини пелюстки (petal_width) для різних видів Iris'
)

fig.update_layout(
    title={
        'x': 0.5,
        'xanchor': 'center'
    }
)

fig.show()
```

3.Побудуйте лінійний графік, що відображає зміну середнього значення "sepal_length" в залежності від значення "sepal_width" для кожного виду квітки (setosa, versicolor, virginica).

```python
import plotly.express as px
import pandas as pd

df = px.data.iris()

df_grouped = df.groupby(['sepal_width', 'species'], as_index=False)['sepal_length'].mean()

fig = px.line(
    df_grouped,
    x='sepal_width',
    y='sepal_length',
    color='species',
    title='Зміна середнього sepal_length залежно від sepal_width для різних видів Iris'
)

fig.update_layout(
    title={
        'x': 0.5,
        'xanchor': 'center'
    }
)

fig.show()
```

2. Для датасету  px.data.gapminder() побудувати наступні графіки

1.Побудувати гістограму, яка відображає розподіл ВВП за країнами для року 2007. Обрати 10 країн з найбільшим ВВП та позначити їх на гістограмі з неперервною зміною кольору. Додати підписи над стовпцями та вирівняти їх горизонтально. Назва графіку має бути розташована за центром.

```python
import plotly.express as px

df = px.data.gapminder()

df_2007 = df[df['year'] == 2007]

df_top10 = df_2007.sort_values(by='gdpPercap', ascending=False).head(10)

fig = px.bar(
    df_top10,
    x='country',
    y='gdpPercap',
    color='gdpPercap',
    text='gdpPercap',
    title='Топ-10 країн за ВВП на душу населення (2007)'
)

fig.update_traces(
    texttemplate='%{text:.0f}',
    textposition='outside'
)

fig.update_layout(
    title={
        'x': 0.5,
        'xanchor': 'center'
    },
    xaxis_tickangle=-45
)

fig.show()
```

2. Побудувати кругову діаграму, яка відображає розподіл ВВП за країнами Європи у 2018 році. Виберіть 5 країн з найбільшим ВВП і підпишіть діаграму за центром, щоб вона була зручною для сприйняття.




```python
import plotly.express as px

df = px.data.gapminder()

df_europe = df[(df['continent'] == 'Europe') & (df['year'] == 2007)]

df_top5 = df_europe.sort_values(by='gdpPercap', ascending=False).head(5)

fig = px.pie(
    df_top5,
    names='country',
    values='gdpPercap',
    title='Топ-5 країн Європи за ВВП на душу населення'
)

fig.update_layout(
    title={'x': 0.5, 'xanchor': 'center'}
)

fig.show()
```

3. Побудувати лінійні графіки для тригонометричних функцій. Назву графіку вирівняти за центром

1/


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 5 * np.pi, 500)

y = 2 * np.cos(x - np.pi/4)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label=r'$y = 2\cos(x - \frac{\pi}{4})$', color='blue', linewidth=2)

plt.title('Графік тригонометричної функції', loc='center', fontsize=14, pad=20)

plt.xlabel('x')
plt.ylabel('y')
plt.grid(True, linestyle='--', alpha=0.7)
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)

plt.xticks([0, np.pi, 2*np.pi, 3*np.pi, 4*np.pi, 5*np.pi],
           ['0', 'π', '2π', '3π', '4π', '5π'])

plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 500)
y = 0.5 * np.sin(2 * x + np.pi / 3)

plt.figure(figsize=(10, 5))
plt.plot(x, y, color='blue', linewidth=2, label=r'$y = \frac{1}{2}\sin(2x + \frac{\pi}{3})$')

plt.title('Графік функції №2', loc='center', fontsize=14)

plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi], ['0', 'π/2', 'π', '3π/2', '2π'])
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2 * np.pi, 2 * np.pi, 2000)
y = 2 * np.tan(0.5 * x + np.pi / 6)

y[np.abs(y) > 20] = np.nan

plt.figure(figsize=(10, 6))
plt.plot(x, y, color='red', linewidth=2, label=r'$y = 2\text{tg}(\frac{1}{2}x + \frac{\pi}{6})$')

plt.title('Графік функції №3', loc='center', fontsize=14)

plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-10, 10)
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi], ['-2π', '-π', '0', 'π', '2π'])
plt.legend()
plt.show()
```
