```python
import numpy as np

#5. Створіть два одновимірних масиви розміром 6, заповніть їх випадковими цілими числами в діапазоні від 1 до 20 та виконайте на них поелементні операції додавання, віднімання та множення.

a = np.random.randint(1, 21, 6)
b = np.random.randint(1, 21, 6)

print("Масив A:", a)
print("Масив B:", b)
print("Додавання A + B:", a + b)
print("Віднімання A - B:", a - b)
print("Множення A * B:", a * b)
```

```python
import numpy as np

# 6. Створіть дві матриці розміром 3x2 та 2x4, заповніть їх випадковими цілими числами в діапазоні від 1 до 5 та перемножте їх між собою.

A = np.random.randint(1, 6, (3, 2))
B = np.random.randint(1, 6, (2, 4))

C = A @ B
print("Матриця A:\n", A)
print("Матриця B:\n", B)
print("Добуток A @ B:\n", C)

```

```python
import numpy as np

# 7.Створіть матрицю розміром 3x3, заповніть її випадковими дійсними числами в діапазоні від 0 до 1 та транспонуйте її.

matrix_3x3 = np.random.rand(3, 3)
transposed_matrix = matrix_3x3.T

print("Оригінальна матриця:\n", matrix_3x3)
print("Транспонована матриця:\n", transposed_matrix)
```

```python
import numpy as np

# 8. Створіть дві матриці розміром 3x3, заповніть їх випадковими цілими числами в діапазоні від 1 до 10 та знайдіть їхню різницю.

A = np.random.randint(1, 11, (3, 3))
B = np.random.randint(1, 11, (3, 3))

difference = A - B

print("Мартиця А:\n", A)
print("Мартиця B:\n", B)
print("Різниця:\n", difference)
```

```python
import numpy as np

# 9. Створіть дві матриці розміром 3x3, заповніть їх випадковими цілими числами в діапазоні від -5 до 5 та знайдіть їх добуток.

A = np.random.randint(-5, 6, (3, 3))
B = np.random.randint(-5, 6, (3, 3))

product = A @ B  # або np.dot(A, B)

print("Мартиця А:\n", A)
print("Мартиця B:\n", B)
print("Добуток A @ B:\n", product)
```

```python
import numpy as np

# 10. Створіть матрицю розміром 3x3, заповніть її випадковими дійсними числами в діапазоні від -1 до 1 та знайдіть вектор-стовпчик, що містить суму елементів кожного рядка матриці.

matrix = np.random.uniform(-1, 1, (3, 3))
row_sums = np.sum(matrix, axis=1 ).reshape(-1, 1)

print("Матриця А:\n", matrix)
print("Вектор стовпчик сум рядків:\n", row_sums)


```
